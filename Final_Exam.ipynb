```python
alldb = pd.read_csv('alldb.csv')
print(alldb.shape)
alldb.head()
```

เลือก Colum ที่ต้องการ

```python
alldb1 = alldb[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey','title','age']]
alldb1
```

```python
alldb1.isnull().any()#เช็คค่าว่าง
```

จัดการ User เป็นตัวเลข 5 หลัก

```python
alldb_User= alldb1[(alldb1['user'].astype(str).str.isnumeric()) & (alldb1['user'].str.len() == 5)]
alldb_User
```

```python
alldb_User.isnull().any()#เช็คค่าว่าง
```

site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_site= alldb_User[(alldb_User['site'].astype(str).str.isnumeric()) & (alldb1['site'].str.len() == 5)]
alldb_site
```

```python
alldb_site.isnull().any() #เช็คค่าว่าง
```

PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น))

```python
alldb_PartitonKey= alldb_site[(alldb_site['PartitionKey'].astype(str).str.isnumeric()) & (alldb1['PartitionKey'].str.len() == 5)]
alldb_PartitonKey
```

```python
alldb_PartitonKey.isnull().any() #เช็คค่าว่าง
```
